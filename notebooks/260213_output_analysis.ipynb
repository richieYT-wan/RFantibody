```python
import Bio
import pandas as pd
import os
import glob
os.getcwd(), glob.glob('../outputs/*/*.csv') 
```

```python
df = pd.concat([pd.read_csv(x) for x in glob.glob('../outputs/*/*.csv')])
df.sample(5)
```

```python
metrics_col = [x for x in df.columns if 'rmsd' in x or 'pae' in x or 'lddt' in x]
df.query('interaction_pae < 10 and (target_aligned_antibody_rmsd < 10 or target_aligned_cdr_rmsd < 10)').sort_values('interaction_pae')[['timestamp', 'framework', 'hotspots', 'rfab_id'] + metrics_col]
```

```python
df.query('interaction_pae > 10 and (target_aligned_antibody_rmsd < 5 or target_aligned_cdr_rmsd < 5)').sort_values('interaction_pae', ascending=False)[['timestamp','framework', 'hotspots', 'rfab_id']+metrics_col]
```

```python
df.to_csv('../outputs/merged_parsed_outputs.csv', index=False)
```
